```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
radius = 1

x = []
y = []
z = []

for angle in np.linspace(-np.pi*0.5, np.pi*0.5, 40):
    xi = np.sin(angle)*radius
    yi = 0
    zi = np.cos(angle)*radius

    point = np.array([np.sin(angle)*radius,
                      0,
                      np.cos(angle)*radius])
    
    z_angle = np.pi*-0.25
    R_z = np.array([[np.cos(z_angle), -np.sin(z_angle), 0],
                 [np.sin(z_angle), np.cos(z_angle), 0],
                 [0,0,1]])
    
    x_angle = np.pi*-0.35
    R_x = np.array([[1, 0, 0],
               [0, np.cos(x_angle), -np.sin(x_angle)],
               [0, np.sin(x_angle), np.cos(x_angle)]])
    
    point = point@R_x
    point = point@R_z

    x.append(point[0])
    y.append(point[1])
    z.append(point[2])
```

```python
plt.scatter(x,z)
plt.scatter(y,z)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, c='b', marker='o')
plt.show()
```
